```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
df=pd.read_csv('imdb_top_1000.csv')
```

```python
df
df.info()
```

```python
# Data processing and cleaning
#Detecting NA values in data set
df.isnull().sum()
```

```python
# Applying this method to see null values
missing_values=[np.nan]
df=pd.read_csv('imdb_top_1000.csv',na_values=missing_values)
```

```python
df.isnull().sum()
```

```python
# to plot null values in a graph
sns.heatmap(df.isnull(),yticklabels=False)
```

```python
# Remove all the NA values
df.dropna(how='all')# having whole row empty
```

```python
# Calculate the median meta_score
median_meta_score = df['Meta_score'].median()

# Fill missing values in the "Meta_score" column with the median Meta_score
df['Meta_score'] = df['Meta_score'].fillna(median_meta_score)
```

```python
# Fill missing values in "Certificate" column using fillna as in dictionary form(As No rating)
df.fillna({
     'Certificate':'No rating'
},inplace=True)
```

```python
# Cheaking data types in the dataframe
df.dtypes
```

```python
#Change data type for 'Gross'
df['Gross'] = df['Gross'].replace({',': ''}, regex=True).astype(float)
#Fill the empty values in Gross column
df['Gross'] = df['Gross'].interpolate(method='linear', limit_direction='both')  # Linear interpolation
df.dtypes
```

```python
#Change data type for'Released_Year'
#Convert 'Released_Year' to numeric, coerce non-numeric values to NaN
df['Released_Year'] = pd.to_numeric(df['Released_Year'], errors='coerce')

# Convert to integer to remove decimal places and handle NaN values
df['Released_Year'] = df['Released_Year'].astype('Int64')
```

```python
#Change data type for 'Runtime'
# Convert 'Runtime' to float and only extrat number(AS value in column is 154 min will extract 154)
df['Runtime'] = df['Runtime'].str.extract('(\d+)').astype('float64')
```

```python
'''----------Analyze how movies directed by different directors perform in terms of gross earnings------------------'''
Director_gross_earning=df.groupby('Director')['Gross'].mean().sort_values(ascending=False)
Director_gross_earning
```

```python
'''---------------Genre popularity over the years.-------------------'''
# Splitting the 'Genre' column into separate genres
df['Genre'] = df['Genre'].str.split(', ')

# Exploding the 'Genre' column to have one row for each genre
df_expanded_genre = df.explode('Genre')
# Counting the occurrences of each genre per year
genre_counts = df_expanded_genre.groupby(['Released_Year','Genre']).size().sort_values(ascending=False).reset_index(name='Genre_Count')
genre_counts
# In 2013 drama has gain populartiy 
```

```python
'''-----------------------Top 10 genre popularity gain--------------------'''
# Finding the top 10 genres based on total popularity across all years
top_genres = genre_counts.groupby('Genre')['Genre_Count'].sum().sort_values(ascending=False).head(10)
top_genres
# Total genre produce generally drama followed by comedy and crime
```

```python
'''---------------Correlation between IMDB ratings and commercial success---------------------'''
# Calculate the correlation between 'IMDB_Rating' and 'Gross'
correlation_with_gross = df['IMDB_Rating'].corr(df['Gross'])
correlation_with_gross
```

```python
 '''---------finding correlation with Meta_sore and IMBD_rating-----------------'''
# Calculate the correlation coefficient
correlation_with_meta_score = df['IMDB_Rating'].corr(df['Meta_score'])
correlation_with_meta_score
```

```python
'''--------------------------Impact of movie length on ratings or earnings-----------------------'''
# Calculate the correlation between 'Runtime' and 'IMDB_Rating'
from scipy.stats import pearsonr
correlation_rating = pearsonr(df['Runtime'], df['IMDB_Rating'])[0]
correlation_rating
```

```python
'''------------------------Own insides------------------------'''
#-------------Which movie is given higest rating-----------------
# Find the movie with the highest rating
highest_rating_index = df['IMDB_Rating'].idxmax()
# Extract the movie details for the highest rating
highest_rating_movie = df.loc[highest_rating_index, 'Series_Title']
print(f"Movie Having higest rating {highest_rating_movie}")
```

```python
#--------------Give top 10 movies with maximum gross profit----------------
# Sort the DataFrame by 'Gross' in descending order and get the top 10
top_10_gross_movies = df.sort_values(by='Gross', ascending=False).head(10)

print("Top 10 movies with maximum gross profit:")
print(top_10_gross_movies[['Series_Title', 'Gross']])
```

```python
#----------------which director has produce maximum movies------------------
# Count the number of movies for each director
director_counts = df['Director'].value_counts()

# Find the director with the maximum number of movies
max_movies_director = director_counts.idxmax()
max_movies_count = director_counts.max()
print(f"The director who has produced the maximum number of movies is {max_movies_director} with {max_movies_count} movies.")
```

```python
#------------------count of movies with different certification-------------------
# Group by certification and count the number of movies for each certification
certification_counts = df.groupby('Certificate')['Series_Title'].count()
certification_counts
```

```python
#-----------------------top 10 movies with maximun number of votes---------------------
# Sort the DataFrame by 'No_of_Votes' in descending order and get the top 10
top_10_votes_movies = df.sort_values(by='No_of_Votes', ascending=False).head(10)

print("Top 10 movies with the maximum number of votes:")
print(top_10_votes_movies[['Series_Title', 'No_of_Votes']])
```

```python
'''------------------Which combination of actors are getting good gross profit ?-----------------------------'''
# for this we need to make new data frame as actor_gross_df which has gross and stars in new df
# Extracting relevant columns
actors_columns = ['Star1', 'Star2', 'Star3', 'Star4']

# Creating a new DataFrame with unique combinations of actors and their total gross
actor_gross_df = pd.DataFrame()
for col in actors_columns:
    actor_gross_df[col] = df[col]
actor_gross_df['Total_Gross'] = df['Gross']


# Grouping by unique combinations of actors and summing the total gross
actor_gross_sum = actor_gross_df.groupby(actors_columns)['Total_Gross'].sum().reset_index()


# Sorting the DataFrame by total gross in descending order to find the top combinations
top_actor_combinations = actor_gross_sum.sort_values(by='Total_Gross', ascending=False).head(10)
top_actor_combinations
```

```python
'''-------------------top 10 Movie with shorest  runtime?---------------------'''
# Find the top 10 movies with the shortest runtime
top_10_shortest_runtime = df.nsmallest(10, 'Runtime')
top_10_shortest_runtime
```

```python
'''--------------top 10 Movie with longest runtime?---------------------'''
# Find the top 10 movies with the longest runtime
top_10_longest_runtime = df.nlargest(10, 'Runtime')
top_10_longest_runtime
```

```python
'''----------------------Top 10 stars work in maximum number of movies-----------------------------'''
# Melt the DataFrame to make it tidy
melted_df = pd.melt(df, id_vars=['Series_Title'], value_vars=['Star1', 'Star2', 'Star3', 'Star4'], var_name='Star_Type', value_name='Star')

# Counting the occurrences of each star
star_counts = melted_df['Star'].value_counts()

# Finding the star with the maximum work
max_work_star = star_counts.head()

print("Star with the maximum work in movies:", max_work_star)
```
